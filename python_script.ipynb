```python
import requests
import json
import pandas as pd

data_dict = {}
id = 0
num = 20
for id in range(0,num):
  res = requests.get('https://randomuser.me/api')
  response = json.loads(res.text)
  results  = response['results']
  for i in range(len(results)):
    data_dict[id] = {}
    data_dict[id]['first_name'] = results[i]['name']['first']
    data_dict[id]['last_name'] = results[i]['name']['last']
    data_dict[id]['gender (actual)'] = results[i]['gender']
  id += 1

actual_df = pd.DataFrame.from_dict(data_dict, orient='index')
actual_df
```

```python
name_list = []
for name in actual_df['first_name']:
  name_list.append(name)

pred_dict = {}
for i in range(len(actual_df)):
  url = 'https://api.genderize.io/?name=' + name_list[i]
  res = requests.get(url)
  response = json.loads(res.text)
  pred_dict[i] = {}
  pred_dict[i]['gender (predict)'] = response['gender']
  pred_dict[i]['probability'] = response['probability']

pred_df = pd.DataFrame.from_dict(pred_dict, orient='index')
pred_df
```

```python
result_df = actual_df.join(pred_df)
result_df['same_gender'] = (result_df['gender (actual)'] == result_df['gender (predict)'])
result_df
```
